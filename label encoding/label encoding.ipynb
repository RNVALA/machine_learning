```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
df=pd.read_csv("homeprices.csv")
df
```

```python
dummies=pd.get_dummies(df.town)
dummies
#this pandas method will give you the dummy variablee
```

```python
#concatinating two dataframe that is df and dummies
merged=pd.concat([df,dummies],axis='columns')
merged
```

```python
#we are dropping the town column we got from dummy varible 
#drop one of this dummy variable column becz of dummy variable trap
#whenever a one variable can be derived from the rest of the variable  this variable are said to be multilcollinear 
#whenvwr you have the multilcollinear it creates a problem of dummpy v trap that can mashup your model
#hence rule of thumb is have to drop one of the dummy variable column
final=merged.drop(['town','west windsor'],axis='columns')
```

```python
final
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
x=final.drop('price',axis='columns')
x
```

```python
y=final.price
y
```

```python
model.fit(x,y)
```

```python
model.predict([[2800,0,1]])
```

```python
model.predict([[3400,0,0]])
```

```python
model.score(x,y)
```
